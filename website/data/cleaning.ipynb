## Cleaning Gun Violence Archive Data
### Mass Shootings

```python
import pandas as pd
import numpy as np

mass_df = pd.read_csv("gunviolencearchive_mass_shooting_all_years.csv")
mass_df.head(10)
```

```python
import dataframe_image as dfi
```

```python
import os

os.makedirs('../recourses/img', exist_ok=True)
dfi.export(mass_df.head(10), '../recourses/img/tables/mass_raw_table.png', table_conversion='chrome')
```

```python
cleaned_mass_df = mass_df.replace(np.nan, '', regex=True)
cleaned_mass_df.drop(columns=['Operations', 'Address'], inplace=True)
cleaned_mass_df['Incident Date'] = pd.to_datetime(cleaned_mass_df['Incident Date']).dt.date
cleaned_mass_df['State'] = cleaned_mass_df['State'].str.title()
cleaned_mass_df['City Or County'] = cleaned_mass_df['City Or County'].str.title()
cleaned_mass_df.head(10)

os.makedirs('../recourses/img', exist_ok=True)
dfi.export(cleaned_mass_df.head(10), '../recourses/img/tables/mass_cleaned_table.png', table_conversion='chrome')
```

## EDA Gun Violence Archive
### Mass Shootings

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data = cleaned_mass_df, x='State', order = cleaned_mass_df['State'].value_counts().index, color='darkred')
plt.xticks(rotation=90)
plt.title('Number of Mass Shootings by State from Jan 2014 to Dec 2024')
plt.xlabel('State')
plt.ylabel('Number of Incidents')
plt.tight_layout()
plt.savefig('../recourses/img/plots/mass_shootings_by_state.png')
plt.show()
```

```python
# Number of Mass Shootings by Year
mass_by_year_df = cleaned_mass_df
mass_by_year_df['Incident Year'] = pd.to_datetime(mass_by_year_df['Incident Date']).dt.year
mass_by_year_df.head(10)

plt.figure(figsize=(10, 6))
sns.countplot(data = mass_by_year_df, x='Incident Year', color='darkred')
plt.title('Number of Mass Shootings by Year from Jan 2014 to Dec 2024')
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.tight_layout()
plt.savefig('../recourses/img/plots/mass_shootings_by_year.png')
plt.show()
```

```python
# pick the columns you care about
cols_of_interest = ["Victims Injured", "Victims Killed", "Suspects Killed", "Suspects Injured", "Suspects Arrested"]

# group by year and sum counts
yearly = mass_by_year_df.groupby("Incident Year")[cols_of_interest].sum().reset_index()

# reshape into long format for seaborn
yearly_melted = yearly.melt(id_vars="Incident Year", 
                            value_vars=cols_of_interest, 
                            var_name="Category", 
                            value_name="Count")

# plot
plt.figure(figsize=(10,6))
sns.barplot(data=yearly_melted, x="Incident Year", y="Count", hue="Category")
plt.title("Victims and Suspects per Year")
plt.xlabel('Year')
plt.ylabel('Number of Incidents')
plt.tight_layout()
plt.savefig('../recourses/img/plots/mass_shootings_injury_killed_by_year.png')
plt.show()
```
